# 119 안전센터의 확장

### DataSet1 : 119 안전센터 1개소당 담당인구

```python
import pandas as pd

station_citizen = pd.read_csv('citizens.csv', sep='\t')
```

```python
station_citizen = station_citizen.drop(['기간'], axis=1)
```

```python
station_citizen
```

### 한글 깨짐 해결

```python
import matplotlib.pyplot as plt

# matplotlib 폰트설정
plt.rc('font', family='NanumGothic') # For Windows
```

```python
station_citizen.info()
```

```python
station_citizen['119안전센터 1개소당 담당인구'] = station_citizen['119안전센터 1개소당 담당인구'].str.replace(',','')
station_citizen['인구'] = station_citizen['인구'].str.replace(',','')
```

```python
station_citizen['119안전센터 1개소당 담당인구'] = station_citizen['119안전센터 1개소당 담당인구'].astype('int64')
station_citizen['인구'] = station_citizen['인구'].astype('int64')

station_citizen
```

```python
station_citizen.corr()
```

```python
y = station_citizen['119안전센터 1개소당 담당인구']
x = station_citizen['자치구']

plt.figure(figsize=(20,5))

plt.bar(x, y, width=0.5, color="blue")

plt.title('119안전센터 1개소당 담당인구', fontsize=20)

plt.xlabel('자치구', fontsize=18)

plt.ylabel('1개소당 담당인구', fontsize=18)

plt.show()
```

#### 평균 이상 출력

```python
station_citizen_upper_mean = station_citizen[station_citizen['119안전센터 1개소당 담당인구']>=83467]
station_citizen_upper_mean
```

### DataSet2 : 자치구별 119구조활동실적 

```python
rescue_gu_list = pd.read_csv('rescue_gu.csv', sep='\t')
rescue_gu_list = rescue_gu_list[['자치구', '119구조활동실적']]

rescue_gu_list= rescue_gu_list.drop([0,1,2,3,29])

rescue_gu_list['119구조활동실적'] = rescue_gu_list['119구조활동실적'].str.replace(',','')

rescue_gu_list['119구조활동실적'] = rescue_gu_list['119구조활동실적'].astype('int64')

rescue_gu_list
```

```python
sum = rescue_gu_list['119구조활동실적'].sum()
rescue_gu_list.loc[30] = ['평균',int(sum/25)]

rescue_gu_list
```

```python
y = rescue_gu_list['119구조활동실적']
x = rescue_gu_list['자치구']

plt.figure(figsize=(20,5))

plt.bar(x, y, width=0.5, color="blue")

plt.title('119안전센터 1개소당 구조활동', fontsize=20)

plt.xlabel('자치구', fontsize=18)

plt.ylabel('1개소당 구조활동', fontsize=18)

plt.show()
```

#### 평균 이상 출력

```python
rescue_gu_list_upper_mean = rescue_gu_list[rescue_gu_list['119구조활동실적']>=6524]
rescue_gu_list_upper_mean
```

### 구별 인구와 119구조활동실적의 상관관계

```python
merged = pd.merge(station_citizen, rescue_gu_list)
merged
```

```python
merged.corr() # 인구와 119구조활동실적은 상관관계가 다소 높다. 0.63
```

#### 119 안전센터의 추가가 필요한 구

```python
intersection_result = pd.merge(station_citizen_upper_mean, rescue_gu_list_upper_mean)
intersection_result
```

```python
intersection_result['119안전센터 1개소당 담당인구 - 평균 인구'] = intersection_result['119안전센터 1개소당 담당인구']-83467
intersection_result['119구조활동실적 - 평균 구조활동실적'] = intersection_result['119구조활동실적']-6524
intersection_result
```
