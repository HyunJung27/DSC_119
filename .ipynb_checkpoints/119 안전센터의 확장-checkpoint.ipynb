# 119 안전센터의 확장

### DataSet1 : 119 안전센터 1개소당 담당인구

```python
import pandas as pd

station_citizen = pd.read_csv('citizens.csv', sep='\t')
```

```python
station_citizen = station_citizen.drop(['기간'], axis=1)
```

```python
station_citizen
```

### 한글 깨짐 해결

```python
import matplotlib.pyplot as plt

# matplotlib 폰트설정
plt.rc('font', family='NanumGothic') # For Windows
```

```python
station_citizen.info()
```

```python
station_citizen['119안전센터 1개소당 담당인구'] = station_citizen['119안전센터 1개소당 담당인구'].str.replace(',','')
```

```python
station_citizen['119안전센터 1개소당 담당인구'] = station_citizen['119안전센터 1개소당 담당인구'].astype('int64')

station_citizen
```

```python
y = station_citizen['119안전센터 1개소당 담당인구']
x = station_citizen['자치구']

plt.figure(figsize=(20,5))

plt.bar(x, y, width=0.5, color="blue")

plt.title('119안전센터 1개소당 담당인구', fontsize=20)

plt.xlabel('자치구', fontsize=18)

plt.ylabel('1개소당 담당인구', fontsize=18)

plt.show()
```

### DataSet2 : 자치구별 119구조활동실적 

```python
rescue_gu_list = pd.read_csv('rescue_gu.csv', sep='\t')
rescue_gu_list = rescue_gu_list[['자치구', '119구조활동실적']]

rescue_gu_list= rescue_gu_list.drop([0,1,2,3,29])

rescue_gu_list['119구조활동실적'] = rescue_gu_list['119구조활동실적'].str.replace(',','')

rescue_gu_list['119구조활동실적'] = rescue_gu_list['119구조활동실적'].astype('int64')

rescue_gu_list
```

```python
sum = rescue_gu_list['119구조활동실적'].sum()
rescue_gu_list.loc[30] = ['평균',int(sum/25)]

rescue_gu_list
```

```python
y = rescue_gu_list['119구조활동실적']
x = rescue_gu_list['자치구']

plt.figure(figsize=(20,5))

plt.bar(x, y, width=0.5, color="blue")

plt.title('119안전센터 1개소당 구조활동', fontsize=20)

plt.xlabel('자치구', fontsize=18)

plt.ylabel('1개소당 구조활동', fontsize=18)

plt.show()
```

### DataSet3 : 소방서별 119구조활동실적

```python
rescue_station_list = pd.read_csv('rescue_station.csv', sep='\t')
rescue_station_list
```

```python
rescue_station_list = rescue_station_list[['소방서', '119구급대활동실적', '119구급대활동실적.1']]

rescue_station_list= rescue_station_list.drop([0,1,2, 27])

rescue_station_list['119구급대활동실적'] = rescue_station_list['119구급대활동실적'].str.replace(',','')
rescue_station_list['119구급대활동실적.1'] = rescue_station_list['119구급대활동실적.1'].str.replace(',','')

rescue_station_list['119구급대활동실적'] = rescue_station_list['119구급대활동실적'].astype('int64')
rescue_station_list['119구급대활동실적.1'] = rescue_station_list['119구급대활동실적.1'].astype('int64')

rescue_station_list.columns=['소방서', '출동건수', '이송건수']

rescue_station_list
```

```python
sum_1 = rescue_station_list['출동건수'].sum()
sum_2 = rescue_station_list['이송건수'].sum()

rescue_station_list.loc[27] = ['평균',int(sum_1/25), int(sum_2/25)]

rescue_station_list
```

```python
y = rescue_station_list['출동건수']
x = rescue_station_list['소방서']

plt.figure(figsize=(20,5))

plt.bar(x, y, label='출동건수', width=0.5, color="blue")

plt.title('119안전센터 1개소당 출동건수', fontsize=20)

plt.xlabel('소방서', fontsize=18)

plt.ylabel('1개소당 출동건수', fontsize=18)

plt.show()
```

```python
y = rescue_station_list['이송건수']
x = rescue_station_list['소방서']

plt.figure(figsize=(20,5))

plt.bar(x, y, width=0.5, color="blue")

plt.title('119안전센터 1개소당 이송건수', fontsize=20)

plt.xlabel('소방서', fontsize=18)

plt.ylabel('1개소당 이송건수', fontsize=18)

plt.show()
```

```python
y1 = rescue_station_list['출동건수']
y2 = rescue_station_list['이송건수']
x = rescue_station_list['소방서']

plt.figure(figsize=(20,5))

plt.bar(x, y1, label='출동건수', width=0.5, color="blue")
plt.bar(x, y2, label='이송건수', width=0.5, color="red")

plt.title('119안전센터 1개소당 출동건수', fontsize=20)

plt.xlabel('소방서', fontsize=18)

plt.ylabel('1개소당 출동건수', fontsize=18)

plt.show()
```
