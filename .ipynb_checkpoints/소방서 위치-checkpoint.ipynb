```python
import geopandas as gpd
import matplotlib.pyplot as plt

plt.rcParams["font.family"] = 'NanumGothic'
plt.rcParams["font.size"] = 13
plt.rcParams["figure.figsize"] = (10,10)
```

```python
seoul_h_dong_gdf = gpd.read_file('seoul_h_dong.shp', encoding='utf-8')
fig, ax = plt.subplots(figsize=(15,15))
seoul_h_dong_gdf.plot(ax=ax)
```

```python
geo_df = gpd.GeoDataFrame(seoul_h_dong_gdf, crs = {'init':'epsg:4326'})
geo_df.plot(ax=ax)
```

```python
location = pd.read_csv('spot119.csv')
crs = {'init':'epsg:4326'}
location
```

```python
geometry = [Point(xy) for xy in zip(location['경도'], location['위도'])]
geometry[:3]
```

```python
geo_df = gpd.GeoDataFrame(location, crs = crs, geometry = geometry)
geo_df.head()
```

```python
fig,ax = plt.subplots(figsize = (15,15))
seoul_h_dong_gdf.plot(ax=ax, color='orange', alpha=0.7, edgecolor='white')
geo_df.plot(ax=ax, markersize=50, color="red", marker="o", label="119 center")
plt.title("서울특별시 자치구 별 119안전센터", fontsize=25)
plt.legend(prop={'size':15})
```
