# 119 안전센터의 확장

```python
import pandas as pd

station_citizen = pd.read_csv('citizens.csv', sep='\t')
```

```python
station_citizen = station_citizen.drop(['기간'], axis=1)
```

```python
station_citizen
```

### 한글 깨짐 해결

```python
import matplotlib.pyplot as plt

# matplotlib 폰트설정
plt.rc('font', family='NanumGothic') # For Windows
```

```python
station_citizen.info()
```

```python
station_citizen['119안전센터 1개소당 담당인구'] = station_citizen['119안전센터 1개소당 담당인구'].str.replace(',','')
```

```python
station_citizen['119안전센터 1개소당 담당인구'] = station_citizen['119안전센터 1개소당 담당인구'].astype('int64')

station_citizen['119안전센터 1개소당 담당인구']
```

```python
y = station_citizen['119안전센터 1개소당 담당인구']
x = station_citizen['자치구']
plt.bar(x, y, width=0.5, color="blue")

plt.title('119안전센터 1개소당 담당인구', fontsize=20)

plt.xlabel('자치구', fontsize=18)

plt.ylabel('1개소당 담당인구', fontsize=18)

fig = plt.figure(figsize=(9,9))

plt.show()
```
